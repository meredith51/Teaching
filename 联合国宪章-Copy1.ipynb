```python
Link=soup.find_all("div",class_="content")[0].find_all("a")   
```

```python
UN = 'http://www.un.org/zh/'
hr=list()
for link in Link:
    hr.append(UN+link.get('href'))
```

```python
hr[2]
```

```python
hr=hr[2:22] #取出有关的地址   确保查找完全。最后是19章，前面有前言和介绍页面
```

```python
hr
```

```python
for i in range(len(hr)):
    hr[i]=hr[i].replace('/../','/')  #网址连接中的多余字符去掉
```

```python
with open("UNCharter.txt", "w") as text_file:
     for k in range(len(hr)):
            r = requests.get(hr[k])
            sp = BeautifulSoup(r.content, 'html.parser')
            s_div=sp.find_all("div",class_="field-item even")
            if k==0:
                s_p=sp.find_all({"h3","p"})
            else:
                s_p=sp.find_all({"h3","p"})
            for i in s_p:
                i=i.strings
                if k==0:
                     i="".join(i).strip('')
                else:
                     i="".join(i)
                print(i)
               
```
